Пример работы программы:

Результат:

```python
year = int(input('Введите год, что бы узнать високосный он или нет: '))
if year % 4 == 0 and year % 100 != 0 or year % 400 == 0:
    print (year, ' - високосный год')
else:
    print (year, ' - не високосный год')
```

Задание 5 (необязательное)
Дана переменная, в которой хранится шестизначное число (номер проездного билета). Напишите программу, которая будет определять, является ли данный билет "счастливым". Билет считается счастливым, если сумма первых трех цифр совпадает с суммой последних трех цифр номера.

Примеры работы программы:

number = 123456
Результат:
Несчастливый билет

number = 123321
Результат:
Счастливый билет

```python
number = int(input())

b = number//1000 

first_num = b // 100 
second_num = b // 10 % 10    
third_num = b % 10

c = number % 1000

fourth_num = c // 100
fifth_num = c // 10 % 10
sixth_num = c % 10

if (first_num + second_num + third_num)==(fourth_num + fifth_num + sixth_num):
    print ('Счастливый билет')
else:
    print ('Несчастливый билет')
```

Задание 6 (необязательное)
Напишите программу, которая сможет вычислять площади трех фигур (круг, треугольник и прямоугольник). Тип фигуры запрашиваем через пользовательский ввод, после чего делаем запрос характеристик фигуры:

если пользователь выбрал круг, запрашиваем его радиус,
если треугольник – длины трех его сторон;
если прямоугольник – длины двух его сторон.
Пример работы программы:

Введите тип фигуры:
Круг

Введите радиус круга:
10
Результат:
Площадь круга: 314.16

Введите тип фигуры:
Треугольник

Введите длину стороны A:
2

Введите длину стороны B:
2

Введите длину стороны C:
3
Результат:
Площадь треугольника: 1.98


### Площадь круга
$$
S = π⋅r^2
$$

### Площадь треугольника:
$$
S=\sqrt[2]{p(p−a)(p−b)(p−c)}\\
p = \frac{(a+b+c)}{2}
$$

### Площадь прямоуголника
$$
S = a ⋅ b
$$

```python
import math
type_fig = input( "Выберите тип фигуры: Круг, Треугольник или Прямоугольник: ") 
if type_fig == "Круг":
    radius = int(input('Введите радиус круга:'))
    print('Площадь круга:', round((math.pi * radius**2),2))
elif type_fig == "Треугольник":
    side_a = int(input('Введите длину стороны A: '))
    side_b = int(input('Введите длину стороны B: '))
    side_c = int(input('Введите длину стороны C: '))
    p = (side_a + side_b + side_c) / 2
    s = math.sqrt(p * (p - side_a) * (p - side_b) * (p - side_c))
    print('Площадь треугольника: ', round(s,2))
elif type_fig == "Прямоугольник":
    side_a = int(input('Введите длину стороны A: '))
    side_b = int(input('Введите длину стороны B: '))
    print('Площадь прямоульника: ', round(side_a * side_b,2))
else:
    print ('Не верно введен тип фигуры')
```
